## Pairs of features

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html
diabities = datasets.load_diabetes(scaled=False)

# Show
# 442 patience, 10 things measured for each patient
diabities.data
```

```python
# Keys
diabities.keys()
```

```python
# Feature names
diabities.feature_names
```

```python
# Blood pressure
# First dimention ":" represents all the values from one number to another number
# If you don't put any numbers, it defaults to: from 0 until len(data), i.e. [0:len(data)], exclusive if the last index
# Second dimention is the forth column located at index 3
bp = diabities.data[:,3]

# Show first 10 (not including 10th index)
bp[:10]
```

```python
# BMI
bmi = diabities.data[:,2]

# Show first 10 (not including index 10)
bmi[:10]
```

```python
# Show data
diabities.data
```

## Relationships

```python
# Create a ffigure and an axis
fig, ax = plt.subplots()

# Scatter plot
ax.scatter(bmi, bp) # default marker is "o"

# Labels
ax.set_xlabel('BMI')
ax.set_ylabel('Blood presssure')
```

## Colours

```python
# Sex values have been normilised in this dataset
sex = diabities.data[:,1]

# Show first 10
sex[:10]
```

```python
# Uniques sex values (should be only two in the original dataset)
np.unique(sex)
```

```python
# Create a ffigure and an axis
fig, ax = plt.subplots()

# Scatter plot
# Change marker
# https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
# "c" is for color. We can set a calar or assign it to a variable for the plot to automatically differentiate colors
ax.scatter(bmi, bp, marker ='.', c = sex)

# Labels
ax.set_xlabel('BMI')
ax.set_ylabel('Blood presssure')
```

## Scale

```python
# Ages were also normilised in this dataset
age = diabities.data[:,0]

# Show first 10
age[:10]
```

```python
# Create a ffigure and an axis
fig, ax = plt.subplots()

# Scatter plot
# Change marker
# https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
# "s" is for scale. It's not very apparent, but some dots here are larger then others representing different ages

ax.scatter(bmi, bp, c = sex, marker ='.', s = age)

# Labels
ax.set_xlabel('BMI')
ax.set_ylabel('Blood presssure')
```
