> NumPy's **polyfit** function fits a polynomial of a specified degree to a given set of data points using least squares regression. It returns the polynomial coefficients in descending order, which can then be used for interpolation, approximation, or trend analysis. ## Data
https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html

```python
# Create x values
# Give me equally spaced x values between 0 and 10
x = np.linspace(0.0, 10.0, 31)

# Create y values
# The relationship between x and y values here is that every y value is 3 times corresponsponding x value
y = 3.0 * x
```

```python
# Create a scatterplo
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x,y)
```

```python
# Use polyfit to fit a line to the data
# m & c are typically used for polyfit. Using them helps to separate values
# Polyfit shows a relationship between m & c

m, c = np.polyfit(x, y, 1)

# Show 
# m is 3.0;
# c is basically 0 because the second value is in e-16, it just represented like this due to some computer algoritm
m, c
```

```python
# Create a scatterplo
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x,y)

# We used informaion that polyfit sent us back: x , m * x + c
# Where x  represents x
# And (m * x + c) represents y
ax.plot(x, m * x + c, color = 'red')
```


https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html

```python
# Create x values
x = np.linspace(0.0, 10.0, 31)

# Create y values
# Add some noise (comment below)
# Here we are using np.random.normal() function to make this dataset to look more like real data, i.e. to add some random numbers that are: 
#   centerred on 0 (mean = 0) and 
#   with std. dev = 0.3 (around 90% of val will be spread between (0 and -0.3) and (0 and 0.3)) 
#   the amount of values (size) is the same as x value
y = 3.0 * x + np.random.normal(0.0, 0.3, size=x.size)
```

```python
# Create a scatterplo
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x,y)
```

```python
# Use polyfit to fit a line to the data
m, c = np.polyfit(x,y,1)
 
# See what m & c came out now
m, c
```

```python
# Create a scatterplot
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x,y)

# We used informaion that polyfit sent us back: x , m * x + c
# Where x  represents x
# And (m * x + c) represents y
ax.plot(x, m * x + c, color = 'red')
```

## All together

```python
# Create x values
# More points (i.e. 300 instead of 31) will make it more representative visually 
x = np.linspace(0.0, 10.0, 31)

# Create y values
# Creating more noise (i.e. 10.0 instead of 1.0) create more spread dataset
y = 3.0 * x + 0.0 + np.random.normal(0.0, 1.0, size=x.size)

# Use polyfit to fit a line to the data
# Polyfit calculation is to look for the best value representing relationship between x & y
#   in a form of the least distance between value dots from a line
m, c = np.polyfit(x,y,1)
 
# See what m & c came out now
m, c

# Create a scatterplot
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x,y)

# We used informaion that polyfit sent us back: x , m * x + c
ax.plot(x, m * x + c, color = 'red')

# Set title 
ax.set_title(f'y = {m: .2f}x + {c:.2f}')
```
