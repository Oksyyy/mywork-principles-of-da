```python
# Numerical arrays
import numpy as np

# Plotting
import matplotlib.pyplot as plt
```

## Simple Linear Regression

$y = mx + c$

$y = 2x + 1$

```python
# An array of x values
x =np.linspace(0.0, 10.0, 21)

# An array of x values, with noise
y = 2.0 * x + 1.0 + np.random.normal(0.0, 1.0, x.size)
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y, marker = '.', label = '$y=2x+1s$')

# Add legend
ax.legend()
```

```python
# Polyfit
np.polyfit(x,y, 1)
```

## Calculation from Wikipedia

https://en.wikipedia.org/wiki/Simple_linear_regression

```python
# x mean
xbar = np.mean(x)

# y mean
ybar = np.mean(y)
```

```python
# Numerator
num = np.sum((x - xbar) * (y - ybar))

# Denominator
den = np.sum((x - xbar)**2)

# Slope
m = num/ den

# Show m 
m
```

```python
# Intercept
c = ybar - (m * xbar)

# Show c
c
```

$ Cost(m) \sum_i (y_i = m x_i)^2$

## Approximation

```python
# x values
x = np.linspace(0.0, 10.0, 21)

# y values
y = 3.0 * x + np.random.normal(0.0, 1.0, x.size)

# Cost function
def cost(m):
    return np.sum((m*x - y) **2)

# Show cost for m = 3.0
cost(3.0)
```

```python
# A range of m values
m = np.linspace(-3.0, 9.0, 101)

# Calculate the cost function for each m
costs = np.array([cost(mi) for mi in m])

# Plot the cost function
fig, ax = plt.subplots()
ax.plot(m, costs)
```

```python
# Optimize m using a trick
guess = 1.0

# Learning rate
alpha = 0.01

# Keep going
keep_going = True

while keep_going:
    if cost(guess + alpha) < cost(guess):
        guess += alpha
    elif cost(guess - alpha) < cost(guess):
        guess -= alpha
    else:
        keep_going = False

# Show guess
guess
```
